```python
# is it true that people who earn more than 50K have at least high school education?
#data.education.value_counts()
pd.crosstab(data.salary,data.education,margins=True) # Verficar como es el tema de at least.
# ---> Se verifica NO SE ENTENDIÓ LA PREGUNTA. Se requeria saber cual era el grado de la gente que tenia >50K

# Solución de Yorko:
# data.loc[data['salary'] == '>50K', 'education'].unique() # No
```

```python
# Display age statistics for each race (race feature) and each gender (sex feature).
# Use groupby() and describe().
# Find the maximum age of men of Amer-Indian-Eskimo race
# data.race.value_counts()

pd.crosstab([data.sex],[data.race],values=data.age,aggfunc=[np.max]) # Maximun age of AIE men is 82 yo
# ---> Se verifica OK

# Solucion de Yorko (larga y no elegante)
# for (race, sex), sub_df in data.groupby(['race', 'sex']):
#    print("Race: {0}, sex: {1}".format(race, sex))
#    print(sub_df['age'].describe())
```

```python
# Among whom is the porportion of those who earn a lot (>50K) greater: married or single man (marital-status feature)?
# Consider as married those who have a marital-status startng with Married (Married-civ-spouse, Married-spouse-absent or
# Married-AF-Spouse), the rest are considered bachelors.
# data['marital-status'].value_counts()
d={'Married-civ-spouse':'Married'
  ,'Married-spouse-absent:':'Married'
  ,'Married-AF-spouse':'Married'
  ,'Never-married':'Bachelor'
  ,'Divorced':'Bachelor'
  ,'Separated':'Bachelor'
  ,'Widowed':'Bachelor'}
data['estadocivil']=data['marital-status'].map(d)
data.head()
```

```python
# Ahora si el resultado.
pd.crosstab(data.salary,data.estadocivil,margins=True,normalize=True) # Answer: Married
#---> Se verifica OK

# Solucion de Yorko (tediosa y larga aunque con codigo interesante):
# data.loc[(data['sex'] == 'Male') &
#     (data['marital-status'].isin(['Never-married', 
#                                   'Separated', 
#                                   'Divorced',
#                                   'Widowed'])), 'salary'].value_counts()
# data.loc[(data['sex'] == 'Male') &
#     (data['marital-status'].str.startswith('Married')), 'salary'].value_counts()
# data['marital-status'].value_counts()
```
